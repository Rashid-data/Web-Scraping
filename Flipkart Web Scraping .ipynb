# import packages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# copy the URL of Webpage

```python
url= "https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DMi"
```

# Parse the content into HTML

```python
req = requests.get(url)
```

```python
content = BeautifulSoup(req.content, "html.parser")
```

```python
print(content)
```

# Rest code

```python
data = content.find_all("div",{"class":"_2kHMtA"})
```

```python
print(data)
```

```python
links=[]
phn_name=[]
start_link="https://www.flipkart.com"
```

```python
for items in data:
    rest_link=items.find("a")["href"]
    name=items.find("div",attrs={"class":"_4rR01T"})
    phn_name.append(name.text)
    links.append(start_link+rest_link)
```

```python
print(phn_name[0])
print(links[0])
```

```python
dataframe={"Phone_names":phn_name,"Links":links}
final_dataframe=pd.DataFrame(dataframe)
print(final_dataframe)
```

```python
final_dataframe.to_csv("fk_data_url.csv")
```
