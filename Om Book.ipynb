# import packages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# copy the URL of Webpage

```python
url= "https://www.ombooksinternational.com/101-Series~106~106~~.html"
```

# Parse the content into HTML

```python
req = requests.get(url)
```

```python
content = BeautifulSoup(req.content, "html.parser")
```

```python
print(content)
```

# Rest code

```python
data = content.find_all("div",{"class":"product product__style--3 col-lg-2 col-md-2 col-sm-6 col-6"})
```

```python
print(data)
```

```python
book_price=[]
book_name=[]
start_link="https://www.ombooksinternational.com"
```

```python
for item in data:
    # Extract book name
    name = item.find("div", attrs={"class": "product__content content--center"})
    book_name.append(name.text.strip())  # strip to remove leading/trailing whitespaces

    # Extract book price
    price = item.find("div", attrs={"class": "prize d-flex"})
    book_price.append(price.text.strip() if price else 'N/A')  # use 'N/A' if price is not found
```

```python
print(book_name[:5])
print(book_price[:5])
```

```python
import csv
```

```python
# Create a CSV file and write the data
csv_file_path = "books_data.csv"
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Book Name", "Book Price"])  # Write header
    writer.writerows(zip(book_name, book_price))  # Write data

print(f"Data has been scraped and saved to {csv_file_path}")
```
